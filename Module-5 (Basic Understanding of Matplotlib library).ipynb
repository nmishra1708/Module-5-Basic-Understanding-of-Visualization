# Basic Understanding about Visualization using "matplotlib library"

##  How to create simple line plot

```python
#Import libraries
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#Preparing Data
a = np.linspace(0,10,20)
b = np.linspace(10,0,20)

#Adding figure
fig = plt.figure(figsize=(10,5))

#Adding axes
ax1 = plt.subplot()

#Simple line plot of both a and b
ax1.plot(a)
ax1.plot(b)

#Show the plot
plt.show()

'''
It is best to use line plot when compairing fewer then 25 numbers. It is quick simple way to represent data.
'''
```

```python
# Customized line plot

#Import libraries
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#Preparing Data
a = np.linspace(0,10,20)
b = np.linspace(10,0,20)

#Adding figure
fig = plt.figure(figsize=(10,5))

#Adding axes
ax1 = plt.subplot()

#Simple line plot of both a and b
# Customized line plot->line width, line style, line color, line opacity and marker options
ax1.plot(a, linewidth=2.0, linestyle="--", color="b", alpha=0.5, marker='o')
ax1.plot(b,linewidth = 1.0, linestyle = "--", color='r', alpha=1, marker='^')

#Customization label
plt.xlabel('x-axis')
plt.ylabel('y-axis')

#Customized legend
plt.legend(['a','b'], loc='best')

#Add grid to the plot
plt.grid(True)

#Save the plot
plt.savefig('Lineplot.png')

#Add the title
plt.title('Show the comparison')

#Show the plot
plt.show()
```

```python
# Use subplotting while comparing plots
# Importig libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Preparing Data 
a=np.linspace(0,10,20)
b=np.linspace(10,0,20)

#Adding Figure
fig = plt.figure(figsize=(10,5))

#Sub Plotting
ax1 = plt.subplot(211) # 2nd row 1st column and 1st position
ax2 = plt.subplot(212) # 2nd row 1st column and 2nd position

#Customized- lineWidth, lineStyle, lineColor, lineOpacity, marker etc.
ax1.plot(a, linewidth=2.0, linestyle="--", color='b', alpha=0.5, marker='o')
ax2.plot(b, linewidth=1.0, linestyle="--", color='g', alpha=1, marker='^')

#Setting title of subplot
ax1.set(title='a')
ax2.set(title='b')

#Adding space between subplot
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.6 )

#Showing plot
plt.show()

'''
Use subplotting for compairing plots.
'''
```

# Create Bar chart

```python
#import library
import matplotlib.pyplot as plt
%matplotlib inline

#Preparing Data
data = {'apples':10,'orange':15,'lemon':5,'limes':20}
names = list(data.keys())
values = list(data.values())

#Add figure
fig = plt.figure(figsize=(10,5))

#subplotting
ax1 = plt.subplot()

#plot
ax1.bar(names, values)

#Customization title
plt.title('bar plot')

#Customization x-axis label and y-axis label
plt.xlabel('Names')
plt.ylabel('values')

#Showing plot
plt.show()

'''
bar chart basically use for represent categorical data with rectangular format according to the given values.
'''
```

```python
#Customized Bar Chart

#import library
import matplotlib.pyplot as plt
%matplotlib inline

#Preparing Data
data = {'apples':10,'orange':15,'lemon':5,'limes':20}
names = list(data.keys())
values = list(data.values())

#Add figure
fig = plt.figure(figsize=(10,5))

#subplotting
ax1 = plt.subplot()

#plot
ax1.bar(names, values, align='center', color='orange')

#Customization title
plt.title('bar plot')

#Customization x-axis label and y-axis label
plt.xlabel('Names')
plt.ylabel('values')

#Customized add grid
plt.grid()

#Showing plot
plt.show()
```

```python
# Create customized scatter plot

# Import library
import matplotlib.pyplot as plt
%matplotlib inline

#Preparing Data
a = [10,20,30,40,50,60,70,80]
b = [6,7,3,4,6,7,8,9]
x = [1,2,3,4,3,2,1,5]

#Add Figure
fig = plt.figure(figsize=(10,5))

#Add axes
ax1 = plt.subplot()

# Plot chart
# Add customized color, size, edge color, marker, alpha
ax1.scatter(a,b, c='g', s=200, edgecolor='y', marker='o', alpha=0.5)
ax1.scatter(a,x, c='r', s=400, edgecolor='b', marker='^',alpha=1)

#Add Customized  title
plt.title("Scatter Plot")

# Add customized grid
plt.grid(True)

#Add legend
plt.legend([a,b], loc='best')

# show the plot
plt.show()
```

# Understanding about Box Plot

```python
#Data Preparation
total = [20,4,5,20,30,40,30,10,30,10]
orders = [10,3,1,15,18,2,30,44,2,1]
discount = [30,20,10,5,20,10,60,40,20,1]

data = list([total, orders, discount])
data
```

```python
# Add Figure
fig = plt.figure(figsize=(7,7))

#add axes
ax1 = plt.subplot()

#plot data
ax1.boxplot(data,showmeans=True)

#Add Title
plt.title('Box Plot')

#Customization add grid
plt.grid(True)

#show plot
plt.show()

'''
Display the summary of dataset in efficient way and also understand outlier analysis.
'''
```

# Understanding Violin plot

```python
#import Library
import matplotlib.pyplot as plt
%matplotlib inline

#Preparing Data
total = [20,4,1,30,20,10,20,70,30,10]
orders = [10,3,1,15,17,2,30,44,2,1]
discount = [30,20,10,5,20,10,60,20,40,1]
data = list([total,orders,discount])

#Add figure
fig = plt.figure(figsize=(7,7))

#Add axes
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)

#violin chart plot
ax1.violinplot(data, showmeans=True, showmedians=False)
ax2.violinplot(data, showmeans=False, showmedians=True)

#Add axes title
ax1.set_title('Violin plot_showing means')
ax2.set_title('Violin plot_showing medians')

#Customized add grid
ax1.grid(True)
ax2.grid(True)

#Show the plot
plt.show()
'''
It allow to visualize the distribution of numeric variable for one or several groups. Adapted when the amount of data is huge.
'''
```

### Quiver Plot -> Show vector lines as arrows, Useful in electrical eng. to visualize electrical potential

```python
# Prepare Data
x_pos = 0
y_pos = 0
x_direct = 1
y_direct =1

#Add Figure
fig = plt.figure(figsize=(10,5))

#add axes
ax1 = plt.subplot()

#Plot
ax1.quiver(x_pos,y_pos,x_direct,y_direct)

#customized title
plt.title('Quiver Plot')

#add grid
plt.grid(True)

#show
plt.show()
```

```python
# Prepare Data
x_pos = [0,0]
y_pos = [0,0]
x_direct = [1,0]
y_direct = [1,-1]

#Add Figure
fig = plt.figure(figsize=(10,5))

#add axes
ax1 = plt.subplot()

#Plot
ax1.quiver(x_pos,y_pos,x_direct,y_direct, scale=5)

#Change the scale limit
ax1.axis([-1.5,1.5,-1.5,1.5])

#customized title
plt.title('Quiver Plot')

#add grid
plt.grid(True)

#show
plt.show()
```

### Stream Plot-> 2D plot used to show the fluid flow and 2D field gradients.

```python
# prepare Data
x = np.arange(0,10)
y = np.arange(0,10)
X,Y = np.meshgrid(x,y)
u = np.ones((10,10))
v = np.zeros((10,10))
```

```python
#add figure
fig = plt.figure(figsize=(10,5))

#add subplot
ax1 = plt.subplot()

#Stream bar plot
ax1.streamplot(X,Y,u,v,density=0.5)

#title
plt.title("Customized stream plot")

#show
plt.show()
```

```python
# Customized stream plot
# prepare Data
x = np.arange(0,2.2,0.1)
y = np.arange(0,2.2,0.1)
X,Y = np.meshgrid(x,y)
u = np.cos(X)*y
v = np.sin(y)*Y

#add figure
fig = plt.figure(figsize=(10,5))

#add subplot
ax1 = plt.subplot()

#Stream bar plot
ax1.streamplot(X,Y,u,v,density=1)

#title
plt.title("Customized stream plot")

#show
plt.show()
```

## Pie Chart->used for represent percentage or proportional data, display 6 categories or fewer.

```python
#prepare data
labels = ['CSE','EC','ME','CE','IT']
size = [60,45,120,90,30]

#Add figure
fig = plt.figure(figsize=(10,5))

#add axes
ax1 = plt.subplot()

#plot
ax1.pie(size, labels=labels, autopct='%1.2f%%',shadow=True,startangle=90)

#title
plt.title('Pie Chart')

#show
plt.show()
```

```python
#prepare data
labels = ['CSE','EC','ME','CE','IT']
size = [60,45,120,90,30]

#add explode
explode=(0,0,0.1,0,0)

#Add figure
fig = plt.figure(figsize=(10,5))

#add axes
ax1 = plt.subplot()

#plot
ax1.pie(size,explode=explode, labels=labels, autopct='%1.2f%%',shadow=True,startangle=90)

#title
plt.title('Pie Chart')

#show
plt.show()
```

## Donut Chart->contain more tehn one data series, each data series that you plot in a donut chart add a ring to chart.

```python
# Prepare Data
# I have 3 group and 7 subgroups
group_names = ['groupA','groupB','groupC']
group_size = [12,11,30]
subgroup_names = ['A.1','A.2','A.3','B.1','B.2','C.1','C.2','C.3','C.4','C.5']
subgroup_size = [4,3,5,6,5,10,5,5,4,6]

#Create Colors
a,b,c=[plt.cm.Blues, plt.cm.Reds, plt.cm.Greens]

#add figure and axes
fig, ax = plt.subplots()
ax.axis('equal')

#plot first ring
mypie1, _=ax.pie(group_size, radius=1.3, labels=group_names, colors=[a(0.6), b(0.6), c(0.6)])

#plot second ring(Inside)
mypie2, _=ax.pie(subgroup_size, radius=1.3-0.3, labels=subgroup_names, labeldistance=0.7, colors=[a(0.5),a(0.4),a(0.3)])

#customize
plt.setp(mypie1, width=0.3, edgecolor='white')
plt.setp(mypie2, width=0.4, edgecolor='white')
plt.margins(0,0)

#show 
plt.show()
```
